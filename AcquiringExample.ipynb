```python
from bayesopt2 import RBFKernel, bayesian_optimisation_1d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.backends.backend_pdf import PdfPages
```

```python
kernel = RBFKernel()
```

```python
fn = lambda x: np.sin(3 * x) + 0.5 * x
```

```python
def spiky_mix(x, noise=0.0, rng=None):
    """
    Many sharp local optima + smooth trend.
    Domain suggestion: [0, 7].
    """
    x = np.asarray(x, dtype=float)
    # Smooth trend + coupled oscillations
    trend = 0.15 * x - 0.03 * (x - 3.0)**2
    waves = 0.9*np.sin(2.8*x) + 0.35*np.cos(7.5*x + 0.3) + 0.2*np.sin(15.0*x + 1.2)

    # A mix of narrow & wide Gaussian bumps/dips
    bumps  = (
        2.7*np.exp(-0.5*((x-0.8)/0.15)**2)  # sharp positive
      - 2.2*np.exp(-0.5*((x-2.1)/0.25)**2)  # sharp negative
      + 1.8*np.exp(-0.5*((x-3.6)/0.35)**2)
      - 1.6*np.exp(-0.5*((x-4.2)/0.18)**2)
      + 2.4*np.exp(-0.5*((x-5.1)/0.50)**2)  # broad positive
      - 1.2*np.exp(-0.5*((x-6.2)/0.22)**2)
    )

    y = trend + waves + bumps

    if noise:
        if rng is None:
            rng = np.random.default_rng()
        y = y + rng.normal(0.0, noise, size=y.shape)
    return y if y.ndim else float(y)
```

```python
obvs, values, data = bayesian_optimisation_1d(spiky_mix, 0, 5, kernel, max_iters=50, acquisition_seeds=20, store_data=True)
```

```python
plt.scatter(obvs, values)
plt.show()
```

```python
def bo_static_steps_to_pdf(
    data,
    pdf_path,
    true_fn=None,
    lower_bound=None,
    upper_bound=None,
    figsize=(8, 6),
    highlight_last=True,
    dpi=130,
):
    """
    Save static BO plots (posterior mean & acquisition at each iteration)
    into a multi-page PDF.

    Parameters
    ----------
    data : dict
        Output from bayesian_optimisation_1d(..., store_data=True).
    pdf_path : str
        Path to the PDF file to create.
    true_fn : callable, optional
        If provided, plot ground-truth function on the top panel.
    lower_bound, upper_bound : float, optional
        x-limits. If None, inferred from X_vals.
    figsize : tuple
        Figure size per page.
    highlight_last : bool
        If True, highlight the newest observation at each step.
    dpi : int
        Resolution used when saving.
    """
    Xg = np.asarray(data["X_vals"])
    means_list  = data["posterior_means"]
    stds_list   = data["posterior_stdevs"]
    acqs_list   = data["acquisition_vals"]
    Xs_list     = data["obvs_locations"]
    Ys_list     = data["obvs_values"]

    n_steps = len(means_list)
    if n_steps == 0:
        raise ValueError("No iterations to plot.")

    if lower_bound is None: lower_bound = float(np.min(Xg))
    if upper_bound is None: upper_bound = float(np.max(Xg))

    f_true = None
    if callable(true_fn):
        f_true = np.array([true_fn(x) for x in Xg])

    with PdfPages(pdf_path) as pdf:
        for i in range(n_steps):
            mu = np.asarray(means_list[i])
            sd = np.asarray(stds_list[i])
            acq = np.asarray(acqs_list[i])
            Xs = np.asarray(Xs_list[i])
            Ys = np.asarray(Ys_list[i])

            fig, (ax_top, ax_bot) = plt.subplots(2, 1, figsize=figsize, sharex=True,
                                                 gridspec_kw={"height_ratios": [2, 1]})

            # --- Top: posterior mean ±1σ and observations
            if f_true is not None:
                ax_top.plot(Xg, f_true, lw=1.2, label="f(x)")

            ax_top.plot(Xg, mu, lw=1.8, label="Posterior mean")
            ax_top.fill_between(Xg, mu - sd, mu + sd, alpha=0.15, label="±1σ")

            if Xs.size:
                ax_top.scatter(Xs, Ys, s=28, zorder=3, label="Observations")
                if highlight_last:
                    ax_top.scatter(Xs[-1], Ys[-1], s=60, zorder=4,
                                   edgecolor="k", linewidth=1.0)

            ax_top.set_xlim(lower_bound, upper_bound)

            # y-lims
            y_min = min(np.min(mu - sd), np.min(Ys) if Ys.size else np.min(mu))
            y_max = max(np.max(mu + sd), np.max(Ys) if Ys.size else np.max(mu))
            pad = 0.05 * (y_max - y_min + 1e-12)
            ax_top.set_ylim(y_min - pad, y_max + pad)

            ax_top.set_title(f"Posterior mean (±1σ) & observations — iter {i+1}/{n_steps}")
            ax_top.legend(loc="best")

            # --- Bottom: acquisition
            ax_bot.plot(Xg, acq, lw=1.5)
            a_min, a_max = float(np.min(acq)), float(np.max(acq))
            pad_a = 0.05 * (a_max - a_min + 1e-12)
            ax_bot.set_xlim(lower_bound, upper_bound)
            ax_bot.set_ylim(a_min - pad_a, a_max + pad_a)
            ax_bot.set_title("Acquisition function")
            ax_bot.set_xlabel("x")

            fig.tight_layout()
            pdf.savefig(fig, dpi=dpi)
            plt.close(fig)  # close to free memory
    print(f"Saved BO progress to {pdf_path}")
```

```python
bo_static_steps_to_pdf(
    data,
    true_fn=spiky_mix,                    # optional
    lower_bound=0, upper_bound=5, pdf_path = 'optimiation2.pdf')
```
