```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Notation:
- X is the domain. This is discretised for the sake of plotting
- $\mu_{domain}$ is just the prior mean. This is set to 0 for the sake of demonstration

```python
X_domain = np.linspace(0, 10, 101)
mu_domain = np.zeros_like(X_domain)
```

The kernel determines the covariance matrix in place. The kernel below is the squared exponential kernel; that is:
$$
k(x_i, x_j) = \sigma^2 * e^{-\frac{\vert x_i - x_j \vert^2}{l ^ 2}}
$$

, Parameterised by the standard deviation $\sigma$ and the length scale $l$

```python
def squared_exponential_kernel_with_grad(x1, x2, length_scale=1.0, stdev=1.0):
    dx = x1[:, None] - x2[None, :]
    sq_dist = dx**2
    K = (stdev**2) * np.exp(-0.5 * sq_dist / (length_scale**2))
    dK_dx1 = K * (-dx) / (length_scale**2)
    return K, dK_dx1

def squared_exponential_kernel(x1, x2, length_scale=1.0, stdev=1.0):
    sq_distance = (x1[:, None] - x2[None, :]) ** 2
    return stdev**2 * np.exp(-0.5 * sq_distance/(length_scale ** 2))
```

```python
K, _ = squared_exponential_kernel_with_grad(X_domain, X_domain, stdev=5)
```

```python
K
```

```python
samples = np.random.multivariate_normal(mean = mu_domain, cov=K, size=100)
```

```python
# Plot the realisations
for item in samples[:10]:
    plt.plot(X_domain, item)

# plot the credible intervals
std_error = np.diag(K) ** 0.5
plt.fill_between(X_domain, mu_domain - 1.96 * std_error, mu_domain + 1.96 * std_error, color = 'blue', alpha=0.2)

plt.show()
```

```python
# Check validity of credible intervals
samples = np.array(samples)
np.sum(np.where(np.abs(samples) < 5 * 1.96))/10**6
```

## **Bayesian Updating**

### **General Case**

Now, the process for updating conducting a Bayesian update on a GP

First, some machinery is introduced.

Given two sets of random variables $\mathbf{X}^{(1)} = (X_1 ...., X_q)$ and $\mathbf{X}^{(2)} = (X_{q+1}, ..... X_p)$, with:

$$
\mathbf{X} = 
\begin{pmatrix}
\mathbf{X}^{(1)} \\
\mathbf{X}^{(2)}
\end{pmatrix}
=
\begin{pmatrix}
X_1 \\
\vdots \\
X_q \\
X_{q+1} \\
\vdots \\
X_p
\end{pmatrix}
$$

, such that $\mathbf{X}^{(1)} \sim \mathcal{N}(\boldsymbol{\mu ^ {(1)}}, \boldsymbol{\Sigma_{11})}$, $\mathbf{X}^{(2)} \sim \mathcal{N}(\boldsymbol{\mu ^ {(2)}}, \boldsymbol{\Sigma_{22})}$, and $\text{Cov}(\mathbf{X}^{(1)}, \mathbf{X}^{(2)}) = \boldsymbol{\Sigma_{12}}$

If $\boldsymbol{\Sigma_{12}} = 0$, then $\mathbf{X}^{(1)} \text{ and } \mathbf{X}^{(2)}$ are uncorrelated, and the marginal distribution of $X_i$ is just a normal distribution with mean $\boldsymbol{\mu_i} \text{ and covariance } \boldsymbol{\Sigma_{ii}}$. This seems reasonable.

This can be extended for correlated $\mathbf{X}^{(1)} \text{ and } \mathbf{X}^{(2)}$. Particularly, introducing $\mathbf{Y}^{(1)} \text{ and } \mathbf{Y}^{(2)}$ such that $\mathbf{Y}^{(1)} \text{ and } \mathbf{Y}^{(2)}$ and uncorrelated by choosing a suitable B in:

$$
\mathbf{Y}^{(1)} = \mathbf{X}^{(1)} + B\mathbf{X}^{(2)}, \mathbf{Y}^{(2)} = \mathbf{X}^{(2)}
$$

Can solve for $B$ in:

$$
\mathbf{0} = \mathbb{E}\!\big[(\mathbf{Y}^{(1)} - \mathbb{E}[\mathbf{Y}^{(1)}])(\mathbf{Y}^{(2)} - \mathbb{E}[\mathbf{Y}^{(2)}])^\top\big]
$$

$$
= \mathbb{E}\!\big[(\mathbf{X}^{(1)} + \mathbf{B}\mathbf{X}^{(2)} - \mathbb{E}[\mathbf{X}^{(1)}] - \mathbf{B}\mathbb{E}[\mathbf{X}^{(2)}])(\mathbf{X}^{(2)} - \mathbb{E}[\mathbf{X}^{(2)}])^\top\big]
$$

$$
= \mathbb{E}\!\big[(\mathbf{X}^{(1)} - \mathbb{E}[\mathbf{X}^{(1)}] + \mathbf{B}(\mathbf{X}^{(2)} - \mathbb{E}[\mathbf{X}^{(2)}]))(\mathbf{X}^{(2)} - \mathbb{E}[\mathbf{X}^{(2)}])^\top\big]
$$

$$
= \mathbf{\Sigma}_{12} + \mathbf{B}\mathbf{\Sigma}_{22}.
$$

, so $B = -\boldsymbol{\Sigma_{12}}\boldsymbol{\Sigma_{22}}^{-1}$

Now, we can solve for the distribution of $\mathbf{Y}$

$$
\mathbb{E}\!\begin{bmatrix} \mathbf{Y}^{(1)} \\ \mathbf{Y}^{(2)} \end{bmatrix}
= \mathbb{E}\!\left[
\begin{bmatrix}
\mathbf{I} & -\mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1} \\
\mathbf{0} & \mathbf{I}
\end{bmatrix}
\mathbf{X}
\right]
$$

$$
= 
\begin{bmatrix}
\mathbf{I} & -\mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1} \\
\mathbf{0} & \mathbf{I}
\end{bmatrix}
\begin{bmatrix}
\boldsymbol{\mu}^{(1)} \\
\boldsymbol{\mu}^{(2)}
\end{bmatrix}
= 
\begin{bmatrix}
\boldsymbol{\mu}^{(1)} - \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\boldsymbol{\mu}^{(2)} \\
\boldsymbol{\mu}^{(2)}
\end{bmatrix}
=
\begin{bmatrix}
\boldsymbol{\nu}^{(1)} \\
\boldsymbol{\nu}^{(2)}
\end{bmatrix}
= \boldsymbol{\nu}
$$

and covariance given by:

$$
\mathbb{E}\!\big[(\mathbf{Y} - \boldsymbol{\nu})(\mathbf{Y} - \boldsymbol{\nu})^\top\big]
=
\begin{bmatrix}
\mathbb{E}[(\mathbf{Y}^{(1)} - \boldsymbol{\nu}^{(1)})(\mathbf{Y}^{(1)} - \boldsymbol{\nu}^{(1)})^\top] &
\mathbb{E}[(\mathbf{Y}^{(1)} - \boldsymbol{\nu}^{(1)})(\mathbf{Y}^{(2)} - \boldsymbol{\nu}^{(2)})^\top] \\[6pt]
\mathbb{E}[(\mathbf{Y}^{(2)} - \boldsymbol{\nu}^{(2)})(\mathbf{Y}^{(1)} - \boldsymbol{\nu}^{(1)})^\top] &
\mathbb{E}[(\mathbf{Y}^{(2)} - \boldsymbol{\nu}^{(2)})(\mathbf{Y}^{(2)} - \boldsymbol{\nu}^{(2)})^\top]
\end{bmatrix}
$$

$$
=
\begin{bmatrix}
\mathbf{\Sigma}_{11} - \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\mathbf{\Sigma}_{21} & \mathbf{0} \\
\mathbf{0} & \mathbf{\Sigma}_{22}
\end{bmatrix}.
$$


As $\mathbf{Y}^{(1)}$ and $\mathbf{Y}^{(2)}$ are uncorrelated, the joint pdf is given by:

$$
\mathcal{N}\!\big(\mathbf{Y}^{(1)} \,\big|\, \boldsymbol{\mu}^{(1)} - \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\boldsymbol{\mu}^{(2)},\;
\mathbf{\Sigma}_{11} - \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\mathbf{\Sigma}_{21}\big)
\;\;
\mathcal{N}\!\big(\mathbf{Y}^{(2)} \,\big|\, \boldsymbol{\mu}^{(2)},\; \mathbf{\Sigma}_{22}\big)
$$

Want to reformulate w.r.t. $\mathbf{X}$.  
Particularly, 
$$
\mathbf{Y}^{(1)} = \mathbf{X}^{(1)} - \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\mathbf{X}^{(2)},
$$
so
$$
\mathbf{X}^{(1)} = \mathbf{Y}^{(1)} + \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\mathbf{X}^{(2)}.
$$

Given the independence established previously, conditioning on $\mathbf{X}^{(2)} = \mathbf{x}^{(2)}$, and taking the expectation of $\mathbf{X}^{(1)}$, we have:

$$
\mathbb{E}[\mathbf{X}^{(1)} \mid \mathbf{X}^{(2)} = \mathbf{x}^{(2)}]
= \boldsymbol{\mu}^{(1)} + \mathbf{\Sigma}_{12}\mathbf{\Sigma}_{22}^{-1}\big(\mathbf{x}^{(2)} - \boldsymbol{\mu}^{(2)}\big).
$$

And the covariance is independent of the realisation of $\mathbf{x}^{(2)}$ in this parameterisation.  
This procedure for determining the posterior can be applied to Bayesian optimisation.


### **For Bayesian Optimisation**

Considering a desire to optimise an objective function, modelled as some Gaussian process. Observations are drawn as $y = f(x) + \epsilon$, where $\epsilon \sim \mathcal{N}(0, \sigma_n^2)$

Given some observation locations $\mathbf{x} = (x_1, x_2 .... x_n)$ and observation values $\mathbf{y} = (y_1, y_2 ..... y_n)$, we desire to determine the posterior $p(f \vert \mathbf{y})$

First, we create the joint $p(f, \mathbf{y})$ as a Gaussian process. The mean of $\begin{bmatrix} f \\ \mathbf{y} \end{bmatrix} $ is just $\begin{bmatrix} \boldsymbol{\mu} \\ \mathbf{m} \end{bmatrix} $, where $\boldsymbol{\mu}$ is just the prior mean on $f$, and $\boldsymbol{m}$ is the mean of $f$ at the observation points.

Determining the covariance matrix present here is slightly more involed. We construct it in block matrix form, with the following:

1. $\boldsymbol{\Sigma}_{ff}$ is the kernel applied to the domain $\mathbf{X}$:  
   $$
   \boldsymbol{\Sigma}_{ff} = k(\mathbf{X}, \mathbf{X})
   $$

2. $\boldsymbol{\Sigma}_{fy}$ is the kernel applied between the domain and the observation points:  
   $$
   k(f(x_i), y_j) = k(f(x_i), f(x_j) + \epsilon_j) 
   = k(f(x_i), f(x_j)) + k(f(x_i), \epsilon_j).
   $$
   With $\epsilon_j \overset{\text{iid}}{\sim} \mathcal{N}(0, \sigma_n^2)$, this reduces to  
   $$
   \boldsymbol{\Sigma}_{fy} = k(\mathbf{X}, \mathbf{X}_{\text{obs}}).
   $$

3. $\boldsymbol{\Sigma}_{yy}$ is the kernel applied to all pairs of observation points:  
   $$
   k(y_i, y_j) = k(f(x_i) + \epsilon_i,\; f(x_j) + \epsilon_j)
   = k(f(x_i), f(x_j)) + k(f(x_i), \epsilon_j) + k(\epsilon_i, f(x_j)) + k(\epsilon_i, \epsilon_j).
   $$
   Given our assumptions on $\epsilon$, this simplifies to  
   $$
   \boldsymbol{\Sigma}_{yy} = k(\mathbf{X}_{\text{obs}}, \mathbf{X}_{\text{obs}}) + \sigma_n^2 \mathbf{I}.
   $$


Therefore, we can specify that:

$$
p(\mathbf{f}, \mathbf{y}) \;=\; \mathcal{GP}\!\left(
\begin{bmatrix}
\mathbf{f} \\[6pt]
\mathbf{y}
\end{bmatrix}
;\;
\begin{bmatrix}
\boldsymbol{\mu} \\[6pt]
\mathbf{m}
\end{bmatrix},
\begin{bmatrix}
\boldsymbol{\Sigma}_{ff} & \boldsymbol{\Sigma}_{fy}^\top \\[6pt]
\boldsymbol{\Sigma}_{fy} & \boldsymbol{\Sigma}_{yy}
\end{bmatrix}
\right)
$$

Applying the standard conditioning rule, we obtain:

$$
p(\mathbf{f} \mid \mathbf{y}) \;=\; \mathcal{GP}\!\left(\mathbf{f};\; \boldsymbol{\mu}_D,\; \boldsymbol{\Sigma}_D\right)
$$

where

$$
\boldsymbol{\mu}_D \;=\; \boldsymbol{\mu} 
+ \boldsymbol{\Sigma}_{fy}^\top \boldsymbol{\Sigma}_{yy}^{-1}\big(\mathbf{y} - \mathbf{m}\big),
$$

$$
\boldsymbol{\Sigma}_D \;=\; \boldsymbol{\Sigma}_{ff}
- \boldsymbol{\Sigma}_{fy}^\top \boldsymbol{\Sigma}_{yy}^{-1}\boldsymbol{\Sigma}_{fy}.
$$


Note that this framework can handle both the noiseless and the noisy case; merely set $\sigma_n$ to zero



## **Acquiring**

### **Noiseless**

In the abscence of noise, can consider expected improvement as given by:

$$
\alpha_{\mathrm{EI}}(x;\mathcal{D}) 
= \mathbb{E}\!\left[ \max\big(0, f(x) - f^\star \big) \;\middle|\; \mathcal{D} \right]
$$



The best value observed so far is
$$
f^\star = \max_{x_i \in \mathcal{D}} f(x_i),
$$

where $\mathcal{D} = \{(x_i, f(x_i))\}_{i=1}^n$ is the set of observations.






Writing this as an integral:
$$
\alpha_{\mathrm{EI}}(x;\mathcal{D})
= \int_{-\infty}^{\infty} \max\!\big(f - f^\star, 0\big)\,
\mathcal{N}\!\big(f;\,\mu_D(x), \sigma^2(x)\big)\, df.
$$

, where $\sigma^2 (x)$ is just the $\Sigma_{ff,xx}$; that is posterior variance at x

Splitting at $f^\star$:
$$
\alpha_{\mathrm{EI}}(x;\mathcal{D})
= \int_{f^\star}^{\infty} f\,\mathcal{N}\!\big(f;\,\mu_D(x),\sigma^2(x)\big)\,df
- f^\star \int_{f^\star}^{\infty} \mathcal{N}\!\big(f;\,\mu_D(x),\sigma^2(x)\big)\,df.
$$

This yields the closed form:
$$
\alpha_{\mathrm{EI}}(x;\mathcal{D})
= \big(\mu_D(x) - f^\star\big)\,
\Phi\!\left(\frac{\mu_D(x) - f^\star}{\sigma(x)}\right)
+ \sigma(x)\,\phi\!\left(\frac{\mu_D(x) - f^\star}{\sigma(x)}\right),
$$
where $\Phi$ and $\phi$ are the standard normal CDF and PDF.




Various methods can be used for optimising $\alpha$ here. Garnett suggests multistart gradient descent

$$
\frac{\partial \alpha_{\mathrm{EI}}}{\partial \mu_D(x)}
= \Phi\!\left(\frac{\mu_D(x) - f^\star}{\sigma(x)}\right),
$$

$$
\frac{\partial \alpha_{\mathrm{EI}}}{\partial \sigma(x)}
= \phi\!\left(\frac{\mu_D(x) - f^\star}{\sigma(x)}\right).
$$

Applying the chain rule gives
$$
\frac{d\alpha_{\mathrm{EI}}(x;\mathcal{D})}{dx}
= \frac{\partial \alpha_{\mathrm{EI}}}{\partial \mu_D(x)} \cdot \frac{d\mu_D(x)}{dx}
+ \frac{\partial \alpha_{\mathrm{EI}}}{\partial \sigma(x)} \cdot \frac{d\sigma(x)}{dx}.
$$

Particularly:

$$
\frac{d\mu_D(x)}{dx} 
= \frac{d}{dx}\,\mu(x) + \frac{\partial}{\partial x}\,\boldsymbol{\Sigma}_{fy}(x)^\top \alpha
$$

Note that $\alpha = \boldsymbol{\Sigma}_{yy}^{-1}(\mathbf{y} - \mathbf{m})$ is independent of $x$, 
and the prior mean is $\mu(x) = 0$.

As such, we have

$$
\frac{d\mu_D(x)}{dx} 
= \left(\frac{\partial}{\partial x}\,\boldsymbol{\Sigma}_{fy}(x)\right)^\top \alpha.
$$


For the 1D RBF kernel
$$
k(x,x_i)=\sigma_f^2 \exp\!\Big(-\frac{(x-x_i)^2}{2\ell^2}\Big),
$$
its derivative w.r.t. the test input $x$ is
$$
\frac{\partial}{\partial x}k(x,x_i)
= k(x,x_i)\,\frac{x_i - x}{\ell^2}.
$$

Therefore, since $\displaystyle \frac{d\mu_D(x)}{dx} 
= \Big(\frac{\partial}{\partial x}\,\boldsymbol{\Sigma}_{fy}(x)\Big)^\top \alpha$,
we have the closed form
$$
\boxed{\;
\frac{d\mu_D(x)}{dx}
= \sum_{i=1}^n \alpha_i\, k(x,x_i)\,\frac{x_i - x}{\ell^2}
\;}
$$
with $\alpha=\boldsymbol{\Sigma}_{yy}^{-1}(\mathbf{y}-\mathbf{m})$ and (here) $\mathbf{m}=\mathbf{0}$.

Below, an implementation of this is found:

```python
# Define the domain -> just for the purposes of plotting

X_domain = np.linspace(0, 10, 101)
mu_domain = np.zeros_like(X_domain)

# Some observation points that we will condition on

X_obs = np.random.uniform(0, 10, 5)
mu_obs = np.zeros_like(X_obs)

sigma_noise = 0
stdev = 0.5 # This is scale parameter of the kernel


Sigma_ff, _ = squared_exponential_kernel_with_grad(X_domain, X_domain, stdev=stdev)
Sigma_yy, _ = squared_exponential_kernel_with_grad(X_obs, X_obs, stdev=stdev)
Sigma_yy = Sigma_yy + np.diag(np.full(len(X_obs), sigma_noise**2))

Sigma_fy_T, dSigma_fy_dx_T = squared_exponential_kernel_with_grad(X_domain, X_obs, stdev=stdev)

# dSigma_fy_dx_t is just the derivative of the Kernel with respect to x

Sigma_fy = Sigma_fy_T.T

dSigma_fy_dx = dSigma_fy_dx_T.T
```

```python
dSigma_fy_dx_T.shape
```

; this is just the derivative wrt at every point of the initial covariance matrix. This is just used for plotting

```python
sample = np.random.multivariate_normal(mu_obs, Sigma_yy) # Take a sample to condition on
```

```python
jitter = 1e-8 # Numerical stability
L = np.linalg.cholesky(Sigma_yy + jitter * np.eye(Sigma_yy.shape[0]))
rhs = (sample - mu_obs)
alpha = np.linalg.solve(L.T, np.linalg.solve(L, rhs))

mu_posterior = mu_domain + Sigma_fy.T @ alpha

V = np.linalg.solve(L, Sigma_fy) # L^-1 \Sigma_fy
cov_posterior = Sigma_ff - V.T @ V 

dmu_dx = dSigma_fy_dx.T @ alpha


dVdx = np.linalg.solve(L, dSigma_fy_dx)
var = np.diag(cov_posterior)
sigma = np.sqrt(np.clip(var, 0.0, None))
dvar = -2.0 * np.sum(V * dVdx, axis=0)
dsigma_dx = 0.5 * dvar/sigma
```

```python

plt.figure(figsize = (12, 5))
plt.plot(X_domain, mu_posterior, label="Posterior Mean", color="red")

# 95% confidence interval
std_posterior = np.sqrt(np.diag(cov_posterior))
plt.fill_between(
    X_domain,
    mu_posterior - 1.96 * std_posterior,
    mu_posterior + 1.96 * std_posterior,
    color="orange",
    alpha=0.3,
    label="95% CI"
)

# Draw samples from the posterior
posterior_samples = np.random.multivariate_normal(mu_posterior, cov_posterior, size=5)
for i, posterior_sample in enumerate(posterior_samples):
    plt.plot(X_domain, posterior_sample, alpha=0.6, label=f"Sample {i+1}")

# Plot observed points (the actual sample from earlier)
plt.scatter(X_obs, sample, color="black", zorder=5, label="Observed Points")

plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Posterior Mean, Samples, and 95% CI")
plt.legend()
plt.show()
```

```python
f_best = np.max(sample)
xi = 0.01
improv = mu_posterior - f_best - xi
z = np.divide(improv, sigma, out=np.zeros_like(improv), where=sigma > 0)

Phi = norm.cdf(z)
phi = norm.pdf(z)

acquisition = improv * Phi + sigma * phi
acquisition[sigma == 0] = 0.0

dEI_dx = Phi * dmu_dx + phi * dsigma_dx
dEI_dx[sigma == 0] = 0.0

# Plot
fig, ax1 = plt.subplots(figsize=(8,4))

color1 = "tab:blue"
ax1.plot(X_domain, acquisition, color=color1, label="EI(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("EI(x)", color=color1)
ax1.tick_params(axis="y", labelcolor=color1)
ax1.grid(True, ls="--", alpha=0.6)

ax2 = ax1.twinx()
color2 = "tab:red"
ax2.plot(X_domain, dEI_dx, color=color2, label="dEI/dx")
ax2.set_ylabel("dEI/dx", color=color2)
ax2.tick_params(axis="y", labelcolor=color2)

fig.tight_layout()
plt.title("Expected Improvement and its derivative")
plt.show()
```
